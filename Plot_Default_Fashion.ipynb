```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
adam_column = pd.read_csv('C://Users//conne//Master Thesis//Final_Experimental_Results//Hyper//Mnist//Adam//Column.csv')
adam_standard = pd.read_csv('C://Users//conne//Master Thesis//Final_Experimental_Results//Hyper//Mnist//Adam//Standard.csv')
sca_column = pd.read_csv('C://Users//conne//Master Thesis//Final_Experimental_Results//Hyper//Mnist//SCA//Column.csv')
sca_standard = pd.read_csv('C://Users//conne//Master Thesis//Final_Experimental_Results//Hyper//Mnist//SCA//Standard.csv')
adam_column = adam_column.iloc[:,1:]
adam_standard = adam_standard.iloc[:,1:]
sca_column = sca_column.iloc[:,1:]
sca_standard = sca_standard.iloc[:,1:]
```

```python
iters = 600
plt.figure(figsize=(6,5))
plt.clf()
plt.xlabel('Iteration')
plt.ylabel('Objective function')
plt.title('Objective function')
plt.plot(range(iters), pd.ewma(adam_column['err'], halflife=10), label='Adam_Column_wise')
plt.plot(range(iters), pd.ewma(sca_column['err'], halflife=10), label='SCA_Column_wise')
plt.plot(range(iters), pd.ewma(adam_standard['err'], halflife=10), label='Adam_Standard')
plt.plot(range(iters), pd.ewma(sca_standard['err'], halflife=10), label='SCA_Standard')
plt.grid()
plt.legend(loc='upper right')
```

```python
Data_Set = pd.DataFrame()
Data_Set['adam_colum'] = adam_column['acc']
Data_Set['sca_colum'] = sca_column['acc']
Data_Set['adam_standard'] = adam_standard['acc']
Data_Set['sca_standard'] = sca_standard['acc']
```

```python
boxplot = Data_Set.boxplot(column=['adam_standard','sca_standard','adam_colum', 'sca_colum' ],showmeans=True,figsize=(6,5))
boxplot.set_ylabel('Test Classification Score %')
boxplot.set_title('Performance Comparison on Mnist ')
```
